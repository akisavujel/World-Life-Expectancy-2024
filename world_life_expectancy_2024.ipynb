```python
# Import the required libraries for the project
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# USing the url of world population review as reference
url = "https://worldpopulationreview.com/country-rankings/life-expectancy-by-country"
url
```

```python
# Using the request library to read url if the response shows 200 then its available and if it shows 400 then its unavailable
response = requests.get(url)
response
```

```python
# Use beautiful soup to display correct formats
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# First find all tables in the url
table = soup.find("table", class_ = "wpr-table min-w-full border-collapse")
```

```python
# th has all the column names
column_names = soup.find_all("th")
```

```python
column_names
```

```python
# Print all the column names
column_names_title = [title.text for title in column_names]
column_names_title
```

```python
# Cleaning unwanted words from column
column_names_title = [
    col.replace("â\x86\x93", "").strip()
    for col in column_names_title
]
column_names_title
```

```python
# Find all row values in column
column_data = table.find_all("tr")
```

```python
# Empty list to append row values
all_rows = []

# Using loop formula to find row data from td
for data in column_data:
    row_data = data.find_all("td")
    individual_row_data = [row.text for row in row_data]
    all_rows.append(individual_row_data)
```

```python
# Create the data frame and append rows in column 
df = pd.DataFrame(all_rows, columns=column_names_title)
df
```

```python
# Drop any useless row
df = df.drop(df[df['Country'].isna()].index).reset_index(drop=True)
df
```

```python
# Check the column data 
df.info()
```

```python
#Drop the useless column
df = df.drop(columns=[''])
df
```
